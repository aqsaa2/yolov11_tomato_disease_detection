```python
from ultralytics import YOLO

model = YOLO('yolo11m.pt') 
model.train(
    data='/kaggle/input/tomato-leaf-diseases/data.yaml',
    epochs=50,
    imgsz=225,            
    batch=16,
    patience=20,
    optimizer='AdamW',
    lr0=0.001,
    hsv_h=0.05, hsv_s=0.25, hsv_v=0.15,  
    degrees=15, shear=10,
    flipud=1.0, fliplr=1.0,
    project='tomato_yolo11',
    name='exp_match_rf'
)
```

```python
from ultralytics import YOLO

# ppreviously trained weights
model = YOLO('/kaggle/working/tomato_yolo11/exp_match_rf2/weights/best.pt')

# Fine-tuning for another 15 epochs with a lower learning rate
model.train(
    data='/kaggle/input/tomato-leaf-diseases/data.yaml',
    epochs=15,               # Fine-tune length
    lr0=0.0005,               # Lower learning rate to avoid overfitting
    optimizer='AdamW',
    project='tomato_yolo11',
    name='exp_match_rf2'
)
```
